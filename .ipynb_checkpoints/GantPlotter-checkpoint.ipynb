```python
import plotly.express as px
import pandas as pd
import json
import numpy as np
import altair as alt
alt.renderers.enable('notebook')
alt.renderers.enable('default')
```

```python
with open('./tmp-times.json') as result_json:
    data = json.load(result_json)
ram = '1536' 
df = pd.DataFrame( columns = ['task_name', 'start_time', 'end_time'])
for task in data['tasks']:
    df = df.append({'task_name': task['name'], 'start_time': task['startTime'][ram], 'end_time': task['finishTime'][ram]}, ignore_index=True)
```

```python
df['task_id'] = df['task_name'] + '_' + df.index.astype(str)
```

```python
df = df.sort_values(by=['start_time'])
```

```python
df
```

```python


alt.Chart(df).mark_bar().encode(
    x='start_time',
    x2='end_time',
    y=alt.Y('task_id', 
            sort={"encoding": "x2", "order": "descending"}), # Custom sorting
)
```
