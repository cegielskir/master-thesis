```python
import plotly.express as px
import pandas as pd
import json
import numpy as np
import altair as alt
alt.renderers.enable('notebook')
alt.renderers.enable('default')
```

```python
with open('./generated-dag.json') as result_json:
    data = json.load(result_json)
ram = '1536' 
df_generated = pd.DataFrame( columns = ['task_name', 'start_time', 'end_time'])
for task in data['tasks']:
    df_generated = df_generated.append({'task_name': task['name'], 'start_time': task['config']['scheduledStartTime'], 'end_time': task['config']['scheduledFinishTime']}, ignore_index=True)
df_generated['task_id'] = df_generated['task_name'] + '_' + df_generated.index.astype(str)
```

```python
with open('./dag-extracted.json') as result_json:
    data = json.load(result_json)
ram = 'sdbcs' 
df = pd.DataFrame( columns = ['task_name', 'start_time', 'end_time'])
for task in data['tasks']:
    df = df.append({'task_name': task['name'], 'start_time': task['startTime'][str(ram)], 'end_time': task['finishTime'][str(ram)]}, ignore_index=True)
df['task_id'] = df['task_name'] + '_' + df.index.astype(str)
```
