```python
new_json = []
with open('./montage-0.35-raw-results-normalized.json') as json_workflow:
    process_number_map = {}
    data = json.load(json_workflow)
for ram in data:
    ram_results = data[ram]
    for ram_result in ram_results:
        new_results = {}
        new_results['ram'] = ram
        new_results['task_id'] = ram_result['task_id']
        new_results['result'] = ram_result['result']
        new_results['workflow'] = 'montage'
        new_results['workflow_size'] = '0.35'
        new_results['iteration'] = ram_result['iteration']
        new_json.append(new_results)
```

```python
with open('./montage-0.35-raw-results-normalized-prepared.json', 'w') as result_json:
    json.dump(new_json, result_json)
```
