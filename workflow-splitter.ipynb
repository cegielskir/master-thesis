```python
import json
```

```python
def append_proccess_to_map(process_number_map, process):
    if(process in process_number_map):
        process_number_map[process] += 1
    else:
        process_number_map[process] = 1
    return process_number_map
```

```python
with open('./data-collection/montage-0_35/workflow.json') as json_workflow:
    process_number_map = {}
    data = json.load(json_workflow)
    for process in data['processes']:
        process_name = process['name']
        process_number_map = append_proccess_to_map(process_number_map, process_name)
        new_workflow = {}
        process['workflow-name'] = 'montage'
        process['workflow-type'] = '0.35'
        file_name = process_name + '_' + str(process_number_map[process_name]) + '.json'
        print(file_name)
        for main_key in data.keys():
            if main_key == 'processes':
                new_workflow[main_key] = [process]
            else:
                new_workflow[main_key] = data[main_key]
        with open('./data-collection/montage-0_35/' + file_name, 'w+') as outfile:
            json.dump(new_workflow, outfile)

```
