```python
id_to_myid_map = {}
task_id = 1
tasks_map = {
    'mProject':1,
    'mDiffFit':1,
    'mBgModel':1,
    'mBackground':1,
    'mAdd':1,
    'mConcatFit':1,
    'mViewer':1,
    'mImgtbl':1
}
with open('./generated-workflow-id.json', 'r') as workflow:
    data = json.load(workflow)
for task in data['tasks']:
    id_to_myid_map[task_id] = task['name'] + '_' + str(tasks_map[task['name']]) + '.json'
    tasks_map[task['name']] = tasks_map[task['name']] + 1
    task_id = task_id + 1
```

```python
id_to_myid_map
```

```python
x = re.findall("Metrics: task: \[object Object\] .*}}", content)
```

```python
new_json = []
for ram in [list(range(1536, 3072 ,64)]):
    f = open('results-' + str(ram), 'r')
    content = f.read()
    x = re.findall("Metrics: task: \[object Object\] .*}}", content)
    for line in x:
        new_results = {}
        try:
            new_results['task_id'] = id_to_myid_map[int(re.search(",\"id\":([0-9]*)", line).group(1))]
            print("NO ERROR")
        except:
            print(line)
            new_results['task_id'] = re.search("\[object Object\] ([a-zA-Z]*)", line).group(1) + "_1.json"
        if re.match(".*lambda start.*", line):
            line = line + " completed successfully."            
        new_results['ram'] = str(ram)
        new_results['result'] = line
        new_results['workflow'] = 'montage'
        new_results['workflow_size'] = '0.35'
        new_json.append(new_results)
```

```python
times = []
for task in new_json:
    if(task['task_id'].startswith("mDiffFit")):
        start_time = re.findall("lambda start: (.*?) ", task['result'])
        end_time = re.findall("lambda end: (.*?) ", task['result'])


        times.append(int(end_time[0]) - int(start_time[0]))
```

```python
sum(times) / len(times)
```

```python
sum(times) / len(times)
```

```python
with open('./montage-hyperflow-final.json', 'w') as result_json:
    json.dump(new_json, result_json)
```
